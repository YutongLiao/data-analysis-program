```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd

headers = {}
headers['User-Agent'] = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'

links=list()
for i in range(1,3):
    url = 'https://sh.lianjia.com/ershoufang/pg' + str(i)
    response = requests.get(url, headers=headers)
    html=BeautifulSoup(response.text,'html5lib')
    
    titles=html.find_all('div', attrs={'class': 'title'})
    
    for j in range(30):
        link=titles[j].find('a').get('href')
        links.append(link)

prices=list()
areas=list()
lifts=list()
for i in links:
    url = i
    response2 = requests.get(url, headers=headers)
    text=response2.text
    p_price='<span class="total">(\S+)</span>'
    p_price = re.compile(p_price)
    price = p_price.findall(text)
    prices.append(price)
    p_area =  '<li><span class="label">建筑面积</span>(\S+)</li>'
    p_area = re.compile(p_area)
    area = p_area.findall(text)
    areas.append(area)
    p_lift = '<li><span class="label">配备电梯</span>(\S+)</li>'
    p_lift = re.compile(p_lift)
    lift = p_lift.findall(text)
    lifts.append(lift)

data1=zip(areas,lifts,prices,links)
data=list(data1)
name = ('TA','L','TP','WL')
result = pd.DataFrame(columns=name,data=data)
result.to_csv('C:/Users/lyt/Desktop/lianjiaresult.csv')
```
