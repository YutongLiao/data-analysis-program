```python
import pandas as pd
from scipy.cluster import hierarchy
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler

path = 'C:/Users/lyt/Desktop/data/consume/industry/2019.xlsx'
data = pd.read_excel(path,index_col = 0)
data = data.fillna(0)
data1 = StandardScaler().fit_transform(data)
Z = hierarchy.linkage(data1,method = 'single',metric = 'euclidean')
plt.subplots(figsize=(20,20))
p = hierarchy.dendrogram(Z,leaf_rotation=90)
plt.title('Distribution of Energy Consumption by Industry')
plt.xlabel('industry index')
plt.ylabel('distance(euclidean)')
plt.savefig('Distribution of Energy Consumption by Industry.png')
```
